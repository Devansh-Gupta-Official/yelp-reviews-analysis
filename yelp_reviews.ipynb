Applying previous three processes to our data

```python
def message_cleaning(message):
    Test_punc_removed = [char   for char in message if char not in string.punctuation]
    Test_punc_removed_join = ''.join(Test_punc_removed)
    Test_punc_removed_join_clean = [word   for word in Test_punc_removed_join.split() if word.lower() not in stopwords.words('english')]
    return Test_punc_removed_join_clean
```

```python
df_clean = df_1_5['text'].apply(message_cleaning)
```

Applying Count Vectorizer to our example   **IMPORTANT**

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vectorizer = CountVectorizer(analyzer=message_cleaning)   #adding this function to the pipeline so it runs everytime the object instantiates
df_cv = vectorizer.fit_transform(df_1_5['text'])
```

```python
print(vectorizer.get_feature_names_out())
```

```python
print(df_cv.toarray())
```

**Training the Model with the whole dataset**

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
classifier = MultinomialNB()
label = df_1_5['stars'].values
```

```python
classifier.fit(df_cv,label)
```

```python
testing_sample1=['amazing food! highly recommended']
testing_sample2=['shit food, made me sick']

testing_sample_cv1 = vectorizer.transform(testing_sample1)
testing_sample_cv2 = vectorizer.transform(testing_sample2)
test_predict1=classifier.predict(testing_sample_cv1)
test_predict2=classifier.predict(testing_sample_cv2)
```

**Dividing data into Training and Test prior to training the model**

```python
from sklearn.model_selection import train_test_split
```

```python
X = df_cv
y = label
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
classifier = MultinomialNB()
classifier.fit(X_train,y_train)
```

**Evaluating the Model**

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
y_pred_train = classifier.predict(X_train)
```

```python
y_pred_train
```

```python
cm1 = confusion_matrix(y_train,y_pred_train)
sns.heatmap(cm1,annot=True)
```

```python
y_pred_test = classifier.predict(X_test)
```

```python
cm2 = confusion_matrix(y_test,y_pred_test)
sns.heatmap(cm2,annot=True)
```

```python
print(classification_report(y_test,y_pred_test))
```
